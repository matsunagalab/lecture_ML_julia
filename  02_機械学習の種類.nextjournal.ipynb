## Boston housing

まず紹介するのはボストン市郊外における地域別の住宅価格データです。506個の町(小さな区画)で様々な特徴量・説明変数が載っていますが、このうちどれが目的変数である住宅価格(MEDV)に影響を与えるのか調べるのに使われます。自動的に住宅価格が予測できれば、不動産の鑑定なしに価格評価ができて便利です。

* CRIM 地域ごとの一人当たりの犯罪率 (町ごとの一人あたり) (説明変数)
* ZN 25,000 平方フィート以上の住宅区画の割合 (説明変数)
* INDUS	小売業以外の商業が占める土地面積の割合 (説明変数)
* CHAS チャールズ川沿いかどうかの変数（川沿い1、それ以外0、説明変数）
* NOX 一酸化窒素濃度（ppgm単位、説明変数）
* RM 1戸あたりの部屋数 (説明変数)
* AGE 1940年以前に建設された物件の割合 (説明変数)
* DIS 地域から5つの雇用施設までの重み付き距離 (説明変数)
* RAD 高速道路へのアクセスしやすさ (説明変数)
* TAX 10,000ドルあたりの固定資産税の割合 (説明変数)
* PTRATIO 生徒と教師の比率 (説明変数)
* B	地域における黒人の割合 (説明変数)
* LSTAT	地域における低所得者の割合 (説明変数)
* MEDV 住宅価格の中央値  (1000ドル単位) **(目的変数)**

[boston.csv](https://nextjournal.com/data/QmWXcYso6G9WxJMjF38E3gNshXYrD1FnNueQa7AyWGnn2f?content-type=text/csv&node-id=35380fb6-ff06-40be-9110-65b3973e32db&filename=boston.csv&node-kind=file)

## Iris

次はアヤメ(iris)という花のデータです。現代の統計学の基礎を気づいたフィッシャーが扱ったデータなのでフィッシャーのアヤメと呼ばれます。アヤメの個体150サンプルについて、以下の4つの特徴量を計測して、その種(3種, setosa, versicolor, virginica)がラベル付けされています。特徴量と種の関係をデータから学習できるか？という問題に用いられます。例えば、ラベルを使って教師あり学習を行って種を分ける境界を探る問題や、敢えてラベルを用いずにクラスタリングを行ってその性能を検証する問題です。そうした機械学習の手法を評価検証する用途で用いられることが多いです。

![iris.png](https://nextjournal.com/data/QmcsTUT7gawLt7FGkvJhLYSug8S6Zi8t4DcfkchqF3dzR9?content-type=image/png&node-id=e82311f5-dd56-4ef9-b9f6-25848372aa8b&filename=iris.png&node-kind=file)

* SepalLength ガクの長さ (cm単位、説明変数)
* SepalWidth ガクの太さ (cm単位、説明変数)
* PetalLength 花弁の長さ (cm単位、説明変数)
* PetalWidth 花弁の太さ (cm単位、説明変数)
* Species 種類 (setosa, versicolor, virginica、ラベル・目的変数)

[iris.csv](https://nextjournal.com/data/QmP6fvLv5D5ozPEcbK1vxmbBua2cVZGs7paMudgjhJox5U?content-type=text/csv&node-id=f4bcc9bc-8666-47b1-8e6f-301f03d0a3ba&filename=iris.csv&node-kind=file)

CSVデータを読み込んで基本的なプロットをしてみます。

```julia
# CSVデータの読み込み
iris = CSV.read("/.nextjournal/data-named/QmP6fvLv5D5ozPEcbK1vxmbBua2cVZGs7paMudgjhJox5U/iris.csv")
```

```julia
# x軸をPetalLength、y軸をPetalWidthでプロット。Speciesで色分けをする
scatter(iris.PetalLength, iris.PetalWidth, group=iris.Species, legend=:topleft) 
xaxis!("Petal length (cm)")
yaxis!("Petal width (cm)")
```

## MNIST

MNISTデータは機械学習の画像認識の分野で有名なデータです。0から9までの数字を手書きした画像データ 60,000枚に対してその答えである数字がラベルされています。画像と数字の関係をデータから学習できるか？という問題によく用いられます。数字の自動認識は、例えば郵便番号からの郵便物仕分けに便利です。これも最近は機械学習の手法を評価検証するのに用いられることが多いです。画像から数字を認識させた場合のエラー率が報告されます。

* 1列目 数字の答え (ラベル、目的変数)
* 2列目〜785列目 28×28ピクセルの白黒画像の濃淡 (説明変数)

データ容量の制限から、ここでは画像データ60,000枚を5,000枚に削減したものを置いておきます。

[mnist_reduced.csv](https://nextjournal.com/data/QmUsVajJ792FPfc9AEqm3C2EdemFfCUTERQpEgModigvjW?content-type=text/csv&node-id=6f529c42-b363-490a-afab-1e3c893e50d0&filename=mnist_reduced.csv&node-kind=file)


```julia
# CSVデータの読み込み、labelは正解の数字を示す。その他のカラムは画像の各ピクセルの濃さ(白黒)に対応する
mnist = CSV.read("/.nextjournal/data-named/QmUsVajJ792FPfc9AEqm3C2EdemFfCUTERQpEgModigvjW/mnist_reduced.csv")

# データが1列目のラベル(数字)以外は画像なので、Arrayへ変換して扱うことにする
mnist = Array(mnist)
```

では試しに30番目のエントリ(30行列目)の画像を表示させてみましょう。heatmap関数で白黒画像の濃淡が高いところを明るくなるように表示できます。reshapeという関数を使って、1行×784列のArrayを、28行×28列のArrayへ変換してheatmap関数にかけてみましょう。画像が左に傾いて横になってしまうのは仕様です。

```julia
heatmap(reshape(mnist[20, 2:end], 28, 28)', yflip=true)
```

どうやら9に見えますね。正解のラベルをチェックしてみましょう。

```julia
mnist[20, 1]
```

確かに9でした。
